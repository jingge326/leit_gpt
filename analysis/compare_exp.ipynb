```python
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
df_scratch_pretrain = pd.read_csv('./draw_scratch_pretrain.csv', sep='\t')
```

```python
# plot lines
plt.rcParams["figure.figsize"] = (10,8)
plt.rcParams.update({'font.size': 13})
plt.plot(df_scratch_pretrain['epochs'], df_scratch_pretrain['scratch with reconstruction'], label = 'scratch with reconstruction')
plt.plot(df_scratch_pretrain['epochs'], df_scratch_pretrain['scratch without reconstruction'], label = 'scratch without reconstruction')
plt.plot(df_scratch_pretrain['epochs'], df_scratch_pretrain['reconstruction + freezed cross-entropy + unfreezed cross-entropy'], label = 'reconstruction + freezed cross-entropy + unfreezed cross-entropy')
plt.legend()
plt.show()
```

```python
df_finetune_options = pd.read_csv('./draw_finetune_options.csv', sep='\t')
```

```python
# plot lines
plt.rcParams["figure.figsize"] = (10,8)
plt.rcParams.update({'font.size': 13})
plt.plot(df_finetune_options['epochs'], df_finetune_options['freezed cross-entropy + unfreezed cross-entropy'], label = 'freezed cross-entropy + unfreezed cross-entropy')
plt.plot(df_finetune_options['epochs'], df_finetune_options['freezed cross-entropy + unfreezed reconstrunction & cross-entropy'], label = 'freezed cross-entropy + unfreezed reconstrunction & cross-entropy')
plt.plot(df_finetune_options['epochs'], df_finetune_options['unfreezed cross-entropy'], label = 'unfreezed cross-entropy')
plt.plot(df_finetune_options['epochs'], df_finetune_options['unfreezed reconstrunction & cross-entropy'], label = 'unfreezed reconstrunction & cross-entropy')
plt.legend()
plt.show()
```

```python
# plot lines
plt.rcParams["figure.figsize"] = (10,8)
plt.rcParams.update({'font.size': 14})
plt.plot(df_finetune_options['epochs'], df_finetune_options['freezed cross-entropy'], label = 'freezed cross-entropy')
plt.plot(df_finetune_options['epochs'], df_finetune_options['freezed cross-entropy + unfreezed cross-entropy'], label = 'freezed cross-entropy + unfreezed cross-entropy')
plt.plot(df_finetune_options['epochs'], df_finetune_options['unfreezed cross-entropy'], label = 'unfreezed cross-entropy')
plt.xlabel("epoch")
plt.ylabel("auc")
plt.legend()
plt.show()
```

```python
df_finetune_options_loss = pd.read_csv('./draw_finetune_options_loss.csv', sep='\t')
# plot lines
plt.rcParams["figure.figsize"] = (10,8)
plt.rcParams.update({'font.size': 14})
plt.plot(df_finetune_options_loss['epochs'], df_finetune_options_loss['freezed cross-entropy'], label = 'freezed cross-entropy')
plt.plot(df_finetune_options_loss['epochs'], df_finetune_options_loss['freezed cross-entropy + unfreezed cross-entropy'], label = 'freezed cross-entropy + unfreezed cross-entropy')
plt.plot(df_finetune_options_loss['epochs'], df_finetune_options_loss['unfreezed cross-entropy'], label = 'unfreezed cross-entropy')
plt.xlabel("epoch")
plt.ylabel("cross-entropy loss")
plt.legend()
plt.show()
```
